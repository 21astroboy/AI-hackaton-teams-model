```python
import os
import pandas as pd
import numpy as np
import random
```

```python
df = pd.read_excel("Данные о хакатонщиках.xlsx")
```

```python
df
```

```python
df = df.drop('id', axis=1)
df = df.drop('Flutter, Numpy, Pandas, С различными библиотеками Python и xml', axis=1)
```

```python
list_of_tests = list()

for i in range(400):
    team = df.sample(n=5)
    list_of_tests.append(team)
```

```python
for i in range(400):
    team1 = list_of_tests[i]
    arr = np.arange(1, 6)
    team1 = pd.DataFrame(team1).set_index(keys=arr)
    back = 0
    front = 0
    design = 0
    ml = 0
    for j in team1['Кем ты хочешь быть на хакатонах? ']:
        if 'Бэкенд' in j:
            back += 1
        if 'Дизайнер' in j:
            design += 1
        if 'Фронтенд' in j:
            front += 1
        if 'Фуллстек' in j:
            back += 1
            front += 1
        if 'ML/DS/AI' in j:
            ml += 1
    team = [{'Backend': back, 'Frontend': front, 'ML/DS/AI': ml, 'Design': design}]
    team = pd.DataFrame(team)
    list_of_tests[i] = team
```

```python
for i in range(len(list_of_tests)):
    team = pd.DataFrame(list_of_tests[i])
    if (int(team['Backend']) == 0) | (int(team['Frontend']) == 0) | (int(team['ML/DS/AI']) == 0) | (
            int(team['Design']) == 0):
        team['verdict'] = 0
    else:
        team['verdict'] = 1
    list_of_tests[i] = team
```

```python
df_merged = pd.concat(list_of_tests, ignore_index=True, sort=False)
```

```python
arr = np.arange(1, 401)
df_merged = pd.DataFrame(df_merged).set_index(arr)
```

```python
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras.layers import Dense
%matplotlib inline
```

```python
hack_competence = ['ML', 'Full', 'Design']
def choose_hack_competence():
    return random.choice(hack_competence)
df_merged['hack competence'] = df_merged.apply(lambda x: choose_hack_competence(), axis=1)
```

```python
def set_ML_verdict(row):
    if row['hack competence'] == 'ML' and row['ML/DS/AI'] < 2:
        return 0
    else:
        return 1
df_merged['verdict'] = df_merged.apply(set_ML_verdict, axis=1)
```

```python
def set_backend_verdict(row):
    if (row['Backend'] < 2 or row['Frontend'] < 2) and row['hack competence'] == 'Full':
        return 0
    else:
        return 1
def set_design_verdict(row):
    if (row['Frontend'] < 2 or row['Design'] < 1) and row['hack competence'] == 'Design':
        return 0
    else:
        return 1
```

```python
df_merged['verdict'] = df_merged.apply(set_backend_verdict, axis=1)
df_merged['verdict'] = df_merged.apply(set_design_verdict, axis=1)
```

```python
df_merged['ML_comp'] = df_merged['hack competence'].apply(lambda x: 1 if x == 'ML' else 0)
df_merged['Design_comp'] = df_merged['hack competence'].apply(lambda x: 1 if x == 'Design' else 0)
df_merged['Full_comp'] = df_merged['hack competence'].apply(lambda x: 1 if x == 'Full' else 0)
```

```python
len(df_merged[df_merged['verdict']==0])
```

```python
df_merged = df_merged.drop('hack competence',axis = 1)
```

```python
df_merged
```

```python
from tensorflow.keras.layers import Dense, Flatten
import keras
from keras.models import Sequential
```

```python
df_merged.columns
```

```python
y_train = df_merged['verdict']
x_train = df_merged.drop('verdict',axis = 1)
```

```python
model = Sequential()
model.add(Dense(7, input_dim=7, activation='relu'))
model.add(Dense(40, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

```python
model.fit(x_train, y_train, epochs=12, batch_size=32, validation_split=0.2)
```

```python
df_merged.columns
```

```python
x_test = np.array([[1, 1, 0, 1, 1, 0, 0]])
predictions = model.predict(x_test)
print("Predicted output:", predictions)
```
